```python
from bs4 import BeautifulSoup
import requests
pname=[]
purl=[]
price=[]
rating=[]
review=[]
count1=0
count2=0
count3=0
count4=0
count5=0
for i in list(range(1,21)):
    URL="https://www.amazon.in/s?k=bags&page={}&crid=2M096C61O4MLT&qid=1653465865&sprefix=ba%2Caps%2C283&ref=sr_pg_{}".format(i,i)
    #  print(URL)
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36','Accept-Language': 'en-US, en;q=0.5'})
    webpage = requests.get(URL,headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    # print(webpage.status_code)
    pnames=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')#for products names list
    for i in pnames:
        count1=count1+1
        pname.append(i.text)
        pname=pname[1:]
    purls=soup.find_all('a',class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal')#for products url list
    for i in purls:
        purl.append(i.text)
        count2=count2+1
    prices=soup.find_all('span',class_='a-price-whole')
    for i in prices:
        price.append(i.text)
        count3=count3+1
    ratings=soup.find_all('i',class_='a-icon a-icon-star-small a-star-small-4 aok-align-bottom')
    for i in ratings:
        rating.append(i.text)
        count4=count4+1
    reviews=soup.find_all('span',class_='a-size-base s-underline-text')
    for i in reviews:
        review.append(i.text)
        count5=count5+1
tlist=[count1,count2,count3,count4,count5]
print(tlist)
a=1000
for i in tlist:
  if (a>i):
    a=i
print(a)
print(count5)
```

```python
import pandas as pd
d=pd.DataFrame()
d['prnames']=pname[slice(a-1)]
d['prurl']=purl[slice(a-1)]
d['prprice']=price[slice(a-1)]
d['prrating']=rating[slice(a-1)]
d['prreviews']=review[slice(a-1)]
d.to_csv('pnames.csv')
d
```
